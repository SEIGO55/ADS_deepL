```python
import random
import matplotlib.pyplot as plt
from collections import Counter

candidates = ["Sunagawa","Tsuzuki-san","Nakamura-san","Miyabe-san","Sakae-san"]

# Running the selection 500 times
selections = [random.choice(candidates) for _ in range(500)]

# Counting the frequency of each candidate being selected
frequency = Counter(selections)

# Finding the candidate with the highest frequency
most_frequent = frequency.most_common(1)[0]

# Creating a histogram
plt.figure(figsize=(10, 6))
plt.bar(frequency.keys(), frequency.values(), color='skyblue')
plt.xlabel('Candidates')
plt.ylabel('Frequency')
plt.title('Histogram of Presentation Selection Frequency')
plt.show()

most_frequent, frequency

```
